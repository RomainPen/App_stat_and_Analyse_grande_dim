```python
import pandas as pd

# options d'affichage
pd.set_option("display.min_rows", 20)
pd.set_option("display.max_column", 200)
```

```python
df = pd.read_csv("./MobilePriceDescription/train.csv")
df
```

## 1/ Encoding

```python
def unique_count(df,range_):
    for i in range_:
        if df.columns[i] == 'price_range':
            print(df.columns[i], " : ", len(df[df.columns[i]].unique()), " ", df[df.columns[i]].unique())
        else :
            print(df.columns[i], " : ", len(df[df.columns[i]].unique()))

unique_count(df, range(0,len(df.axes[1])))
```

```python
def dual_value(df,range_):
    for i in range_:
        if len(df[df.columns[i]].unique()) < 3:
            print(df.columns[i], " : ", df[df.columns[i]].unique())
            
dual_value(df, range(0,len(df.axes[1])))
```

```python
def no_dual_value(df,range_):
    list_ = []
    for i in range_:
        if len(df[df.columns[i]].unique()) > 2 and df.columns[i] != 'price_range':
            list_.append(df.columns[i])
    return list_

list_columns_encoding = no_dual_value(df, range(0,len(df.axes[1])))
list_columns_encoding
```

```python
from sklearn.preprocessing import RobustScaler, PowerTransformer, StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df[list_columns_encoding]=ss.fit_transform(df[list_columns_encoding])
df
```

## 3/ Modeling

```python
from sklearn.model_selection import train_test_split, cross_val_score

x = df.drop(['price_range'], axis=1)
y = df['price_range']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

### Ridge classifier 

Généralement utilisé pour des outputs binaire (0, 1)

```python
from sklearn.linear_model import RidgeClassifier
Ridge = RidgeClassifier(random_state=5).fit(x_train,y_train)
Ridge.score(x_test, y_test)
```

```python
X_trains_df=pd.DataFrame(x_train,columns=x_train.columns)
from sklearn.feature_selection import RFE
svc_lin=RidgeClassifier(random_state=5)
svm_rfe_model=RFE(estimator=svc_lin)
svm_rfe_model_fit=svm_rfe_model.fit(X_trains_df,y_train)
feat_index = pd.Series(data = svm_rfe_model_fit.ranking_, index = x_train.columns)
signi_feat_rfe = feat_index[feat_index==1].index
print('Significant features from RFE',signi_feat_rfe)
```

```python
best_feature = list(signi_feat_rfe)
display(type(best_feature))
best_feature
```

```python
X_trains_new=x_train[best_feature]
X_tests_new=x_test[best_feature]

rfe_Ridge=RidgeClassifier(random_state=5)
rfe_Ridge=rfe_Ridge.fit(X_trains_new, y_train)
rfe_Ridge.score(X_tests_new, y_test) 
```

### Ridge Classifier/ Optimisation des hyper-paramètres (voir méthode de laurent + faire graph des évolution de l'accuracy en fonction des hyper-param) 

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import cross_val_score
val_score = []
for alpha in np.arange(0,50, 1):
    score = cross_val_score(RidgeClassifier(alpha=alpha, random_state=5), X_trains_new, y_train, cv=5).mean()
    val_score.append(score)

plt.figure(figsize=(10, 7))
plt.plot(val_score)
```

```python
#fine-tuner les paramètres :
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV

tuned_parameters = {
    'alpha': np.arange(0,50, 1), # learning rate
    'fit_intercept': [True, False], # logistic regression,
    'tol' : [1e-4, 0.0002, 0.0004, 0.0006, 0.0008, 1e-3, 0.002, 0.003, 0.004, 0.005, 0.006, 0.007, 0.008, 0.009, 1e-2, 1e-1, 1e0, 1e1],
    'class_weight' : [None, 'balanced'],
    'solver' : ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga', 'lbfgs'],
    'positive' : [True, False]
}

grid = GridSearchCV(RidgeClassifier(random_state=5),tuned_parameters, n_jobs=-1) 
grid.fit(X_trains_new,y_train)
print(grid.best_params_, grid.best_score_)
```

```python
rfe_Ridge=RidgeClassifier(alpha=0, class_weight='balanced', fit_intercept=False, positive= False,  solver= 'lsqr', tol= 1.0,random_state=5)
rfe_Ridge=rfe_Ridge.fit(X_trains_new, y_train)
rfe_Ridge.score(X_tests_new, y_test) 
```

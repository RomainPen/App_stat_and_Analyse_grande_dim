```python
import pandas as pd

# options d'affichage
pd.set_option("display.min_rows", 20)
pd.set_option("display.max_column", 200)
```

```python
df = pd.read_csv("./MobilePriceDescription/train.csv")
df
```

## 1/ Encoding

```python
def unique_count(df,range_):
    for i in range_:
        if df.columns[i] == 'price_range':
            print(df.columns[i], " : ", len(df[df.columns[i]].unique()), " ", df[df.columns[i]].unique())
        else :
            print(df.columns[i], " : ", len(df[df.columns[i]].unique()))

unique_count(df, range(0,len(df.axes[1])))
```

```python
def dual_value(df,range_):
    for i in range_:
        if len(df[df.columns[i]].unique()) < 3:
            print(df.columns[i], " : ", df[df.columns[i]].unique())
            
dual_value(df, range(0,len(df.axes[1])))
```

```python
def no_dual_value(df,range_):
    list_ = []
    for i in range_:
        if len(df[df.columns[i]].unique()) > 2 and df.columns[i] != 'price_range':
            list_.append(df.columns[i])
    return list_

list_columns_encoding = no_dual_value(df, range(0,len(df.axes[1])))
list_columns_encoding
```

```python
from sklearn.preprocessing import RobustScaler, PowerTransformer, StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df[list_columns_encoding]=ss.fit_transform(df[list_columns_encoding])
df
```

## 3/ Modeling

```python
from sklearn.model_selection import train_test_split, cross_val_score

x = df.drop(['price_range'], axis=1)
y = df['price_range']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

### Naiv Bayes classifier 

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB().fit(x_train,y_train)
GNB.score(x_test, y_test)
```

### GNB Classifier/ Optimisation des hyper-paramètres (voir méthode de laurent + faire graph des évolution de l'accuracy en fonction des hyper-param) 

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import cross_val_score
val_score = []
for var_smoothing in [1e-9, 1e-8, 1e-7, 1e-6, 1e-5, 1e-4,0.0002, 0.0004, 0.0006, 0.0008, 1e-3, 0.002, 0.003, 0.004, 0.005,  0.006, 0.007, 0.008, 0.009, 1e-2, 1e-1, 1e0, 1e1] :
    score = cross_val_score(GaussianNB(var_smoothing=var_smoothing), x_train, y_train, cv=5).mean()
    val_score.append(score)

plt.figure(figsize=(10, 7))
plt.plot(val_score)
```

```python
#fine-tuner les paramètres :
import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV

tuned_parameters = {
    'var_smoothing' : [1e-9, 1e-8, 1e-7, 1e-6, 1e-5, 1e-4,0.0002, 0.0004, 0.0006, 0.0008, 1e-3, 0.002, 0.003, 0.004, 0.005, 0.006, 0.007, 0.008, 0.009, 1e-2, 1e-1, 1e0, 1e1]
}

grid = GridSearchCV(GaussianNB(),tuned_parameters, n_jobs=-1) 
grid.fit(x_train,y_train)
print(grid.best_params_, grid.best_score_)
```

```python
GNB=GaussianNB(var_smoothing=0.1)
GNB=GNB.fit(x_train, y_train)
GNB.score(x_test, y_test) 
```
